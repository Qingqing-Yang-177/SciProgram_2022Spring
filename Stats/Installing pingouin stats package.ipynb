
### One of the libraries we will use for running statistical tests is Pingouin

To participate in in-class exercises this week (and do this week's homework) you will need to install it.

You should go to you terminal or your Anaconda Prompt and run the following command followed by Enter or Return:

conda install -c conda-forge pingouin

**Note that you will want a "clean terminal window" that doesn't already have jupyter notebook running in it.**

If your terminal looks like the following it means that it is already tied up with a jupyter session. In this case you will want to go to the Shell dropdown menu and choose "New Window" and then execute the conda install command there.
