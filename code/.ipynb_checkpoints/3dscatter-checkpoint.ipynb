```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# make three columns of worth of data

d1 = np.random.randint(100, size=100)
d2 = np.random.randint(100, size=100)
d3 = np.random.randint(100, size=100)
```

```python
# put it in a dataframe

df = pd.DataFrame({'d1': d1, 
                  'd2': d2, 
                  'd3': d3})
df
```

```python
# use matplotlib to scatter in 3d

# set the size of the figure
fig = plt.figure(figsize=(12, 12))

# add a 3d subplot the figure object
ax = fig.add_subplot(projection='3d')

# Make the 3d scatterplot giving data in order of x,y,z:
ax.scatter(df['d1'], df['d2'], df['d3'])


# add some labels
# ax is the axis object returned above that points to the subplot
ax.set_xlabel('d1', fontsize=20)
ax.set_ylabel('d1', fontsize=20)
ax.set_zlabel('d3', fontsize=20)

ax.set_title('3D scatter', fontsize=25)

plt.show()


# save it to the current directory
fig.savefig('3dscatter.jpg')
```

```python
# If you have your data in individual arrays rather than dataframe columns
# you can just pass those as the x,y,z


# set the size of the figure
fig = plt.figure(figsize=(12, 12))

# add a 3d subplot the figure object
ax = fig.add_subplot(projection='3d')

# Make the 3d scatterplot giving data in order of x,y,z:
ax.scatter(d1, d2, d3)


# add some labels
# ax is the axis object returned above that points to the subplot
ax.set_xlabel('d1', fontsize=20)
ax.set_ylabel('d1', fontsize=20)
ax.set_zlabel('d3', fontsize=20)

ax.set_title('3D scatter', fontsize=25)

plt.show()
```
