### Quality Check

Within this dataset, you will also notice some NaN values. We have previously seen how to use dropna, but when you think about a data analysis pipeline you need to decide if:
1. drop the trial
2. replace the NaN value
3. drop the participant

This is a broader question that we won't address right now, but if we have a metric for attention or task engagement, that might facilitate this decision. Often researchers will include attention checks to counteract this issue, especially for online tasks. 

For the purposes of this class, we will have a `threshold` rule to decide if we are dropping trials or excluding the participant. However, we can make a threshold rule, i.e., if the number of NaN trials is larger than 8 (25%), we will exclude this participant.

````python
df2.head()
````

````python
df2.key_resp
````

````python
df2.key_resp.isna()
# is not a number
````

````python
df2.key_resp.isna().sum()
````

````python
#number of NaN values in the series
nan_count = df2.key_resp.isna().sum()

#setting the threshold at 25%
threshold=32*0.25

if nan_count>threshold:
    print('consider dropping this participant')
else:
    df2.dropna(inplace=True)
    df2=df2.reset_index(drop=True)
    print("quality check complete for participant")
````
