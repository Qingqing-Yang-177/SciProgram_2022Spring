## Parameters

Functions can take different types of "parameters", which are essentially variables that are created at the start of the execution of the function.  This is helpful for making more abstract functions that perform some computation with their inputs.  

Let's "refactor" our code to make the ramp function more flexible.

For example, in the last version of the ramp it printed out a ramp leading to four asterisks. What if we wanted to be able to print up to some arbitrary size? One approach is to write a function that takes an input that gives the total number of stars we want to get to.

`````python
'a'*3
`````

`````python
def flexible_ramp(n_stars):
    
    """
    Make a ramp of asterisk up to the number of stars.
    
    n_stars: a positive integer controlling how many asterisks get printed
    """
    
    # loop over the values starting from 1 and going up to the
    # value stored in n_stars
    
    for n in range(1, n_stars+1):
        
        # multiplying a string times a number will repeat that string
        print('*'*n)
    
`````

Now we can request a ramp of arbitrary size:

`````python
n= 50

flexible_ramp(n)
`````

We can extend this so the function takes arguments that control the size of the ramp (how many asterisks it goes up to) as well as how many times the ramp gets repeated.

`````python
def flexible_ramp(n_stars, n_loops):
    
    """
    Make a ramp of asterisk up to the number of stars.
    
    n_stars: a positive integer controlling how many asterisks get printed
    n_loops: how many times to repeate the ramp
    """
    
    # loop over the values starting from 1 and going up to the
    # value stored in n_stars
    
    for _ in range(0, n_loops):
    
        for n in range(1, n_stars+1):

            # multiplying a string times a number will repeat that string
            print('*'*n)
`````

`````python
n_stars = 3
n_loops = 3
flexible_ramp(n_stars, n_loops)
`````

## Return values

The functions we have considered so far simply print something out and then finish.  

Sometimes you might like to have your function give back one or more results for further processing by other parts of your program.  

For instance, instead of printing out the sum of x and y, we can redefine the add() function to return the sum using a special keyword `return`:

`````python
# define add() with a return value
def add(x, y):
    
    s = x+y
    
    return s
`````

Now when we run this function, instead of printing out a message, the value of the sum is calculated and passed back and we can collect into a variable for further use.

`````python
add(1,2)
`````

`````python
# check if add() total is less than some treshold
threshold = 10

z = add(7,2)

if z < threshold:
    print('lower')
    
`````

We want a function that tells us whether two input numbers are "equal" or close to equal up to some tolerable difference. This is useful when the numeric precision of the computer gives us numbers that are close to but not exactly some expected value.

The function should take in three parameters: two numbers to compare and a maximum difference between them below which we will say they are equal.

`````python
import math

math.isclose(0.000001, 0,  rel_tol=.000001)
`````

`````python
def close_enough(x, y, max_diff):
    """
    Check if two numbers are close enough for our purposes. Return True if yes.
    
    x: number, etc
    y: number,
    
    returns: Boolean
    """

    
    diff = abs(x-y)
    
    return diff < max_diff
    
    
`````

### Returning multiple values from a function

You can return multiple values from a function if you want.  

For example, we could create a function called `arithmetic` that does a number of operations to x and y and returns sum, difference, and product of the two numbers.

`````python
# define arithmetic() function
def arithmetic(var1, var2):
    
    s = add(var1, var2)
    d = var1-var2
    p = var1*var2
    
    return s, d, p
`````

This new function gives back a `tuple` (similar to a list) that contains all of the return values in one variable.

`````python
results = arithmetic(3, 4)
results
`````

Elements of a tuple can be accessed with square bracket indexing:

`````python
l = [7, -1, 12]
l
`````

`````python
results[2]
`````

You can also unpack the return values into their own variables like this:

`````python
mysum, mydiff, myproduct = arithmetic(3, 4)
print(f'{mysum=}')

print(f'{mydiff=}')

print(f'{myproduct=}')
`````

## Optional parameters with default values

You will sometimes encounter a slightly different function syntax where some of the input values are specified as key-value pairs, or keyword arguments

`````python
def raise_power(x, power=2):
    return x**power
`````

`````python
# one input argument
raise_power(3)
`````

`````python
# with optional input argument
raise_power(3, power=3)
`````

In functions defined like that, the named argument (`power`) is optional because it has a default value as specified in the function definition. 

When we call the function and only pass in an x value, the function calculates $x^2$ using the default power value in the `x**power` line.

In the second example we included an argument for power (4) and in this case the function replaces the default value of 2 and uses our input of 4.

This can be useful if you want a function to have flexibility but you usually use it with some default values. This saves you from having to pass in all of the arguments, and instead you only need to do so if you're doing something outside of the default.

We have seen this in many places already, like when we include optional `delimiter=` argument to pd.read_csv() or optional `hue=` arguments to seaborn plots.

### Positional and key word arguments

The previous raise_power() example had one _positional_ argument and one _keyword_ argument with a default value.

A function can be defined to be entirely positional arguments, entirely keyword arguments, or a mix.

There are a few things about these different kinds of arguments that you should keep in mind when you are writing and using functions.

**Positional arguments are required**: 

Any positional arguments with no default values _have_ to be input to the function, otherwise you'll get a TypeError saying you are missing a required positional argument:

`````python
def add(x, y, print_results=False):
    r = x+y
    
    if print_results:
        print(r)
        
    return r
    
`````

`````python
add(3, print_results=True)
`````

**Positional arguments are assigned in the order they are input** 

If you have a function like this:


```python
def my_division(val1, val2):
    
    return val1/val2
```

Whatever goes in the first position will get assigned to val1 regardless of its name.

`````python
def my_division(val1, val2):

    return val1/val2
`````

`````python
val1 = 99
val2 = 3

my_division(val2, val1)
`````

**When there is a mix of positional and keyword arguments the positional arguments go first**


Writing a function with a mix of argument types should look like this:

```python
def flexible_ramp(n_stars, n_loops=1):
    for i in range(0, n_loops):
        
        for n in range(1, n_stars+1):
            print('*'*n)
```

and not like this with the keyword argument first:

```python
def flexible_ramp(n_loops=1, n_stars):
    for i in range(0, n_loops):
        
        for n in range(1, n_stars+1):
            print('*'*n)
```

Python will throw a SyntaxError in this second case and tell you what the problem was:

`````python
def flexible_ramp(n_loops=1, n_stars):
    for i in range(0, n_loops):

        for n in range(1, n_stars+1):
            print('*'*n)
`````

### When should I use positional versus keyword arguments in a function definition?

Generally it's a good idea to use keyword arguments. There are several benefits to doing this.

One is that it makes the code "self-documented", meaning that the code sort of tells you what it's doing.

Consider these two examples:

```python
sns.relplot(df['height'], df['weight'], df['group'])
```


```python
sns.relplot(y='height', x='weight', hue='group', data=df)
```

In the second example, the role of each input variable is exposed to us in the code itself and as long as know something about the relplot() function we know how height, weight, and group are going to be used in the plot.

Using keyword arguments also loosens the requirement that you remember exactly what order things need to go into a function.


```python
def my_division(a,b):
    return a/b
```


```python
def my_division(numerator=0, denominator=0):
    return numerator/denominator
```

In the second example the function takes care of assigning the inputs to their internal function role using the human-readable argument names and we don't need to remember whether the my_division() function does first argument divided by the second or the other way around.



### What about keyword arguments _without_ default values

A disadvantage of keyword arguments is that they have default values, so if the user does not pass an input to the function the function will proceed anyway. This is not always desirable.

For example, in our last example with `my_division(numerator=0, denominator=0)` I chose some default values so we could have named arguments, but it's not obvious what the default numerator or denominator value should be or if we want one at all.

One solution to this is to use the python special `None` type. This is a special kind of object in Python that occupies a variable but is essentially a non-existent value and many operations will complain if they are given a None.

`````python
a = None
type(a)
`````

`````python
# try to add a None to some number:
a+3
`````

`````python
# try to get the length of a None variable:
len(a)
`````

Using None as the default value for named arguments is a way to have the benefits of using keyword arguments in functions without worrying (most of the time) whether the code is going to execute with some default value that doesn't make sense.

`````python
def my_division(numerator=None, denominator=None):
    return numerator/denominator
`````
